```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
import time

model = YOLO(".\\runs\\detect\\train3\\weights\\best.pt") 

video_path = ".\Demo2.mp4"
cap = cv2.VideoCapture(video_path)

fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
output_file = 'output_video.mp4' 
fps = 20
frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
out = cv2.VideoWriter(output_file, fourcc, fps, frame_size)

class_names = {
    0: 'Cotton',
    1: 'Denim',
    2: 'Leather',
    3: 'Linen',
    4: 'Nylon',
    5: 'Silk'
}

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    for result in results:
        boxes = result.boxes
        for box in boxes:
            xyxy = box.xyxy[0].cpu().numpy()  
            conf = box.conf[0].item()           
            cls = box.cls[0].item()             

            # bounding box
            cv2.rectangle(frame, (int(xyxy[0]), int(xyxy[1])), (int(xyxy[2]), int(xyxy[3])), (255, 0, 0), 2)
            
            class_name = class_names.get(int(cls),'Unknown')
            text = f'Class: {class_name}, Conf: {conf:.2f}'
            font_scale = 0.5
            font_thickness = 2
            text_size = cv2.getTextSize(text, cv2.FONT_HERSHEY_SIMPLEX, font_scale, font_thickness)[0]

            box_x1, box_y1, box_x2, box_y2 = map(int, xyxy)

            # center position for text box
            text_x = box_x1 + (box_x2 - box_x1 - text_size[0]) // 2
            text_y = box_y1 + (box_y2 - box_y1 + text_size[1]) // 2 

            cv2.rectangle(frame, (text_x, text_y - text_size[1] - 5), (text_x + text_size[0], text_y + 5), (255, 255, 255), -1)

            # put text
            cv2.putText(frame, text, (text_x, text_y), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 0, 0), font_thickness)

    # output video
    out.write(frame)
    time.sleep(0.2)

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Output video saved as {output_file}")
```
